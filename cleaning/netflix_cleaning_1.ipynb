```python
import pandas as pd
df = pd.read_csv(r"D:\DBA\Netflix dataset\Dataset\netflix_titles.csv\netflix_titles.csv")
df.head()
```

```python
# checking datatypes.
df.dtypes
```

```python
# changing the date columns.(this is the only datatype anomaly we have in the dataset).

df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
```

```python
# checking null values.
df.isnull().mean()*100
```

```python
df[df['director'].isnull()]
```

```python
df.loc[df['title'] == 'Kota Factory', 'director'] = 'Raghav Subbu'
```

```python
df.drop_duplicates()
```

```python
# dropping all null rows if every data field of the row or index is null.
df.dropna(how = 'all')
```

```python
# replacing the null values of the director, cast and country columns.

df[['director', 'cast' ,'country']] = df[['director', 'cast' ,'country']].fillna('unknown')
```

```python
df.dtypes
```

```python
df.isnull().mean()*100
```

```python
# flagging remaining null value columns.
df['date_added_null'] = df['date_added'].isna().astype(int)
df['rating_null'] = df['rating'].isna().astype(int)
df['duration_null'] = df['duration'].isna().astype(int)
```

```python
df.head()
```

```python
df[df['date_added'].isnull()]
```

```python
# saving the file.
df.to_csv(r'D:\DBA\Netflix dataset\Dataset\Cleaned_netflix.csv', index = False, encoding = 'utf-8')
print('saved!')
```
